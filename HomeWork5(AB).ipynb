### Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при alpha = 5%. Дайте краткие рекомендации команде .Результаты: 1)Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550 . 2) Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228 3) Конверсии : conv1 = 1.05% conv2 = 1.47% .


### Для оценки статистической значимости различий между конверсиями используем Z-критерий, поскольку:

1. распределение нормальное
2. выборки независимые
3. данные количественные

```python
import numpy as np
import scipy.stats as stats

n1 = 15550
n2 = 15550
conv1 = 164 / n1
conv2 = 228 / n2
p_hat = (conv1 * n1 + conv2 * n2) / (n1 + n2)
se = np.sqrt(p_hat * (1 - p_hat) * (1 / n1 + 1 / n2))
z_stat = (conv1 - conv2) / se
p_value = stats.norm.sf(abs(z_stat)) * 2
p_value
```

р-value < alpha (0,05), поэтому мы отвергаем нулевую гипотезу, говорящую о том, что конверсии в 2х анализируемых группах не отличаютсяя.
Таким образом, делаем вывод, что упрощение формы заказа в магазине Утконос привело к увеличению конверсии в покупку.

### Сравниваем метрику конверсия в покупку. Размер выборки - 10000 элементов в каждой группе . Какой статистический критерий тут лучше всего подойдёт и почему ?

Для сравнения метрики конверсии в покупку с размером выборки 10000 элементов в каждой группе, лучше всего подойдет Z-тест. <br>
Z-тест используется для сравнения двух долей и позволяет оценить статистическую значимость разницы между долями. Этот тест дает достаточно точные результаты, когда выборки достаточно большие.
