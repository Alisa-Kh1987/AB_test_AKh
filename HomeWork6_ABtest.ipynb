## Task 1
### Продакт на главной mail.ru решил протестировать в рекомендательной ленте контента вместо карточек со статьями видеоплеер с короткими видео. Нынешний таймспент на юзера в день в среднем равен 25 минут, а стандартная ошибка (SD) равна 156. Мы предполагаем, что в новой версии таймспент на юзера в день изменится на 10%. Средний трафик 20000 человек в день. Посчитайте сколько дней необходимо держать эксперимент при alpha = 5% и beta = 20% .

#### 1. Рассчитаем размер выборки

Нынешний таймспент - 25 мин./день <br>
Предполагаемый таймспент = 25*1,1 = 27,5 <br>
Standard Deviation: 156 <br>

Расчет произведем в калькуляторе (https://statulator.com/SampleSize/ss2M.html)

Предполагая, что объединенное стандартное отклонение составляет 156 единиц, для исследования потребуется выборка размером:
61124 для каждой группы (т.е. общий размер выборки 122248, при условии равных размеров групп), для достижения степени 80% и уровня значимости 5% (двусторонний), для обнаружения истинной разницы в средних между тестом и контрольной группой, равной 2,5 (т.е. 27,5 - 25) единиц.

#### 2. Рассчитаем длительность эксперимента

122248/20000 = 6,1124. <br>
Округляем в большую сторону, длительность равна 7 дн.

## Task 2

Мы предполагаем, что при оптимизации сайта она изменится минимум на 20%.Средний трафик 4000 человек в день. Посчитайте сколько нам нужно дней держать эксперимент при alpha = 5% и beta = 20%

#### 1. Рассчитаем размер выборки

Для расчета используем калькулятор (https://www.evanmiller.org/ab-testing/sample-size.html)

Для проведения тестирования потребуется 592 чел. на один вариант. <br>
Общий размер выборки составит 1184 чел.

#### 2. Рассчитаем длительность проведения тестирования

1184/4000 = 0,296 дн. <br>
Округляем в большую сторону. Продолжительность тестирования составит 1 день.
